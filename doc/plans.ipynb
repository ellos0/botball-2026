# Plans
This document outlines our plans, calculations, and other accumulated data. The objectives in the "Objectives" column should be written in the format: Location/Object. The "Required Action" column lists what the robot needs to accomplish in order to meet the criteria of the objectives listed, thus granting the points listed. The cell(s) at the bottom calculate the points, per objective, per required action.

    <tr>

        <td>Upper Start Box/Poms</td>

        <td>End robot in Upper Start Box.</td>

```python
# Calculate points

class Point:
    def __init__(self,name,value):
        self.name = name
        self.value = value

def add_quantity(x,m):
    print(f"Accounted for quantity of: {x.name}, with quantity: {m.name}.")
    x.value *= m.value
    print(f"Output: {x.value}.")

def add_multiplier(x, m):
    print(f"Added multiplier: {m.name} to: {x.name}")
    x.value *= m.value

def sum_points(x,m):
    print("Calculating sum...")
    toSum = []
    for item in m:
        toSum.append(item.value)
    x.value += sum(toSum)

def end_program(x):
    print("")
    print(f"Finished Calculations. Output: {x.value}")

def define_program(name):
    print(f"Beginning program to calculate points for: {name}")
    print("")

def main():
    define_program("Robot 1 Points")
    
    PomBasic = Point(name="Upper Start Box Pom", value=2)
    PomTotal = Point(name="Pom Output", value=12)
    
    ConeBasic = Point(name="Cone", value=100)
    ConeTotal = Point(name="Cone Output", value=2)

    CleanDeckTotal = Point(name="Clean Deck", value=100)

    RobotMultiplier = Point(name="Robot Multiplier", value=3)
    
    add_quantity(PomTotal, PomBasic)
    add_quantity(ConeTotal, ConeBasic)

    out = Point("Total Points", value=0)
    
    sum_points(out, [PomTotal, ConeTotal])
    add_multiplier(out, RobotMultiplier)
    sum_points(out, [CleanDeckTotal])

    end_program(out)
    
main()
```
